```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("wdi_data.csv", index_col=0)
df.head()
```

```python
df_2019 = df.query("Year == 2019")
df_2019.head()
```

```python
df_2019 = df_2019.query("indicator in ['NY.GDP.PCAP.PP.CD', 'SP.DYN.LE00.IN']")
df_2019.sample(5)
```

```python
df_2019 = df_2019.drop(columns=['Year'])
```

```python
df_pivot = df_2019.pivot( index=['Country Code', 'Country Name', 'Region', 'Income Group', 'Lending Type'], columns='indicator', values='value').reset_index()
df_pivot.head()
```

```python
fig = px.scatter(
    df_pivot,
    x = 'NY.GDP.PCAP.PP.CD',
    y = 'SP.DYN.LE00.IN',
    title = "2019 Health vs. Wealth"
)

fig.show()
```
