```python
import pandas as pd
import missingno as msno
import plotly as plty
import plotly.graph_objects as go
import plotly.express as px

print("Pandas version: ", pd.__version__)
print("missingno version: ", msno.__version__)
print("plotly version: ", plty.__version__)
```

```python
df = pd.read_csv('/content/wdi_data.csv', index_col=0)

df.head()
```

```python
msno.matrix(df)
```

```python
df_ind = df[df["Country Code"] == "IND"]

df_ind
```

```python
df.shape
```

```python
df_ind_gdppcap = df_ind.query("indicator == 'NY.GDP.PCAP.PP.CD'")
df_ind_gdppcap
```

```python
fig = go.Figure()

trace = go.Bar(
    x = df_ind_gdppcap['Year'],
    y = df_ind_gdppcap['value'],
    name = 'India'
)

fig.add_trace(trace)

fig.update_layout(
    title = "GDP over the years",
    xaxis = {"title" : "Year"},
    yaxis = {"title" : "GDP per Capita"}
)

fig.show()
```

```python
fig = px.bar(
    data_frame = df_ind_gdppcap,
    x = "Year",
    y = "value",
    title = "GDP over the years",
    labels = {"value" : "GDP per Capita"}
)

fig.show()
```
